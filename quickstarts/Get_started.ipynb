```
# prompt: label 2 as normal, 3 as overweight, 4 as obese and 1 as underweight from BMI_Z_Score variable

def categorize_bmi(bmi_score):
  if bmi_score == 1:
    return "underweight"
  elif bmi_score == 2:
    return "normal"
  elif bmi_score == 3:
    return "overweight"
  elif bmi_score == 4:
    return "obese"
  else:
    return "unknown" # Handle other values as needed

df['BMI_Category'] = df['BMI_Z_Score'].apply(categorize_bmi)
```
